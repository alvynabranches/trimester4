```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import optimizers
from tensorflow.keras import losses

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score

from warnings import filterwarnings
filterwarnings('ignore')
```

```python
df = pd.read_csv('./data/mushroom/agaricus-lepiota.data', names=['class', 'cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor', 'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color', 'stalk-shape', 'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring', 'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number', 'ring-type', 'spore-print-color', 'population', 'habitat'])
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    df[col] = df[col].apply(lambda x: x if x != '?' else np.nan)
    df[col] = df[col].astype('category')
    df[col] = df[col].cat.codes
```

```python
df.dropna(inplace=True)
```

```python
df.shape[0]
```

```python
df.head(10)
```

```python
X = df.iloc[:,df.columns != 'class'].values
y = df.iloc[:,df.columns == 'class'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
X_train, X_validatation, y_train, y_validation = train_test_split(X_train, y_train, test_size=0.1)
```

```python
X_train.shape
```

```python
model = Sequential()

# input layer
model.add(BatchNormalization(input_shape=X_train.shape[1:]))
model.add(Dense(22, activation='sigmoid'))

# second layer
model.add(Dense(44, activation='sigmoid'))
model.add(Dropout(.2))

# third layer
model.add(Dense(22, activation='sigmoid'))
model.add(Dropout(.2))

# fourth layer
model.add(Dense(11, activation='sigmoid'))

# output layer
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer=optimizers.Adam(learning_rate=1e-3), loss=losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

```python
%%time
history = model.fit(x=X_train, y=y_train, validation_data=(X_validatation, y_validation), epochs=10)
```

```python
model.save('model.h5')
```

```python
model = load_model('model.h5')
```

```python
confusion_matrix(y_test, model.predict(X_test).round(0))
```

```python
accuracy_score(y_test, model.predict(X_test).round(0))
```
